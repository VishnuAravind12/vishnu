```java
public class Matrix {
    //defining member variables
    private int[][] data;
    private int rows;
    private int cols;

    // Constructor
    public Matrix(int[][] initialValues) {
        this.rows = initialValues.length;
        this.cols = initialValues[0].length;
        this.data = initialValues;
    }

    // Setter to set value at a specific position
    public void setValue(int row, int col, int value) {
        data[row][col] = value;
    }

    // Getter to get value from a specific position
    public int getValue(int row, int col) {
        return data[row][col];
    }

    // Method to multiply this matrix with another matrix
    public Matrix multiply(Matrix other) {
        if (this.cols != other.rows) {
            throw new IllegalArgumentException("Matrix dimensions are invalid (first columns has to equal second rows).");
        }

        Matrix result = new Matrix(this.rows, other.cols);

        for (int i = 0; i < this.rows; i++) {   //i is the rows of first matrix
            for (int j = 0; j < other.cols; j++) {   //j is the columns of the second matrix
                for (int k = 0; k < this.cols; k++) {   //k is first columns and second rows
                    result.data[i][j] += this.data[i][k] * other.data[k][j];   //in first, we go right, in second, we go down
                }
            }
        }

        return result;
    }

    // Method to display the matrix
    public void display() {
        for (int i = 0; i < rows; i++) {
            for (int j = 0; j < cols; j++) {
                System.out.print(data[i][j] + " ");
            }
            System.out.println();
        }
    }

    public static void main(String[] args) {
        Matrix matrix1 = new Matrix(2, 2);
        Matrix matrix2 = new Matrix(2, 2);

        // Set values for matrix1
        matrix1.setValue(0, 0, 1);
        matrix1.setValue(0, 1, 2);
        matrix1.setValue(0, 2, 3);
        matrix1.setValue(1, 0, 4);

        // Set values for matrix2
        matrix2.setValue(0, 0, 7);
        matrix2.setValue(0, 1, 8);
        matrix2.setValue(1, 0, 9);
        matrix2.setValue(1, 1, 10);

        // Multiply matrices
        Matrix result = matrix1.multiply(matrix2);

        // Display result
        result.display();
    }
}
```
